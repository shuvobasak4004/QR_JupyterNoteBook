```python
!pip install qrcode ipywidgets
from google.colab import files
import qrcode
import ipywidgets as widgets

# Function to generate QR code from text
def generate_qr_code_from_text(text):
    # Initialize a QRCode object
    qr_code = qrcode.QRCode(
        version=1,
        error_correction=qrcode.constants.ERROR_CORRECT_H,
        box_size=10,
        border=4,
    )

    # Add the text to the QR code
    qr_code.add_data(text)

    # Generate the QR code image
    qr_image = qr_code.make_image(fill_color="black", back_color="white")

    # Save the QR code image
    qr_code_path = "qr_code.png"
    qr_image.save(qr_code_path)

    # Download the QR code image
    files.download(qr_code_path)

# Create the text input widget
text_input = widgets.Textarea(placeholder='Enter the text', description='Text:')

# Create the generate button widget
generate_button = widgets.Button(description='Generate QR Code')

# Function to handle button click event
def generate_button_click(b):
    text = text_input.value
    generate_qr_code_from_text(text)

# Attach the button click event
generate_button.on_click(generate_button_click)

# Display the widgets
display(text_input, generate_button)
```
